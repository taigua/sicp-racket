### 1.2.2  Tree Recursion

````racket
(define (fib n)
  (cond [(= n 0) 0]
        [(= n 1) 1]
        [else (+ (fib (- n 1))
                 (fib (- n 2)))]))
````

````racket
(define (fib n)
  (fib-iter 1 0 n))

(define (fib-iter a b count)
  (if (= count 0)
      b
      (fib-iter (+ a b) a (- count 1))))
````

````racket
(define (count-change amount)
  (cc amount 5))
(define (cc amount kinds-of-coins)
  (cond [(= amount 0) 1]
        [(or (< amount 0) (= kinds-of-coins 0)) 0]
        [else (+ (cc amount
                     (- kinds-of-coins 1))
                 (cc (- amount
                        (first-denomination kinds-of-coins))
                     kinds-of-coins))]))
(define (first-denomination kinds-of-coins)
  (cond [(= kinds-of-coins 1) 1]
        [(= kinds-of-coins 2) 5]
        [(= kinds-of-coins 3) 10]
        [(= kinds-of-coins 4) 25]
        [(= kinds-of-coins 5) 50]))
````

````racket
(count-change 100)
````

#### Exercise 1.13
**Base case:**

For $n = 0$:

$Fib(0) = 0$ and $\frac{\phi^0 - \psi^0}{\sqrt 5} = 0$.

For $n = 1$:

$Fib(1) = 1$ and $\frac{\phi - \psi}{\sqrt 5} = \frac{\frac{1 + \sqrt 5 - (1 - \sqrt 5)}{2}}{\sqrt 5} = 1$.

**Induction:**

Suppose $Fib(n) = \frac{\phi^n - \psi^n}{\sqrt 5}$, then

$$
\begin{flalign}
Fib(n + 1) &= Fib(n) + Fib(n-1) && \\\nonumber
&=  \frac{\phi^n - \psi^n}{\sqrt 5} + \frac{\phi^{n-1} - \psi^{n-1}}{\sqrt 5} &&\\\nonumber
&= \frac{\phi^n + \phi^{n-1} - (\psi^n + \psi^{n-1})}{\sqrt 5} &&\\\nonumber
&= \frac{\phi^{n-1}(1 + \phi) - \psi^{n-1}(1 + \psi)}{\sqrt 5} &&\\\nonumber
\end{flalign}
$$

And
$$
\begin{flalign}
\phi^2 &= \frac{(1 + \sqrt 5)^2}{2^2} &&\\\nonumber
&= \frac{1 + 2\sqrt 5 + 5}{4} &&\\\nonumber
&= \frac{6 + 2\sqrt 5}{4} &&\\\nonumber
&= \frac{3 + \sqrt 5}{2} &&\\\nonumber
&= 1 + \frac{1 + \sqrt 5}{2} &&\\\nonumber
&= 1 + \phi
\end{flalign}
$$

$$
\begin{flalign}
\psi^2 &= \frac{(1 - \sqrt 5)^2}{2^2} &&\\\nonumber
&= \frac{1 - 2\sqrt 5 + 5}{4} &&\\\nonumber
&= \frac{6 - 2\sqrt 5}{4} &&\\\nonumber
&= \frac{3 - \sqrt 5}{2} &&\\\nonumber
&= 1 + \frac{1 - \sqrt 5}{2} &&\\\nonumber
&= 1 + \psi
\end{flalign}
$$

So
$$
\begin{flalign}
Fib(n + 1) &= \frac{\phi^{n-1}(1 + \phi) - \psi^{n-1}(1 + \psi)}{\sqrt 5} &&\\\nonumber
&= \frac{\phi^{n-1}\phi^2 - \psi^{n-1}\psi^2}{\sqrt 5} &&\\\nonumber
&= \frac{\phi^{n+1} - \psi^{n+1}}{\sqrt 5}
\end{flalign}
$$

According to mathematical induction, we have: $Fib(n) = \frac{\phi^n - \psi^n}{\sqrt 5}$.

In order to prove $Fib(n)$ is the closest integer to $\phi^n / \sqrt 5$, we need prove:

$\left| Fib(n) - \frac{\phi^n}{\sqrt 5} \right| < \frac{1}{2}$

Prove:
$$
\begin{flalign}
\left| Fib(n) - \frac{\phi^n}{\sqrt 5} \right| &= \left| \frac{\phi^n -\psi^n}{\sqrt 5} - \frac{\phi^5}{\sqrt 5} \right| &&\\\nonumber
&= \left| - \frac{\psi^n}{\sqrt 5} \right| &&\\\nonumber
&= \left| \frac{\psi^n}{\sqrt 5} \right| &&\\\nonumber
&= \frac{|\psi|^n}{\sqrt 5}
\end{flalign}
$$

In order to prove $\frac{|\psi|^n}{\sqrt 5} < \frac 1 2$, we need to prove $|\phi|^n < \frac {\sqrt 5} 2$.

And $|\phi| < 1$, so $|\phi|^n < 1$. And $\frac {\sqrt 5} 2 > 1$, so:

$\frac{|\psi|^n}{\sqrt 5} < \frac 1 2$

QED.

#### Exercise 1.15

````racket
(define (cube x) (* x x x))
(define (p x) (- (* 3 x) (* 4 (cube x))))
(define (sine theta)
  (if (<= (abs theta) 0.1)
      theta
      (p (sine (/ theta 3.0)))))
````

##### a) The evaluation of `(sin 12.15)`:
```scheme
(sine 12.15)
-> (p (sine 4.05))
-> (p (p (sine 1.35)))
-> (p (p (p (sine 0.45))))
-> (p (p (p (p (sine 0.15)))))
-> (p (p (p (p (p (sine 0.05))))))
```

So the `p` is applied 5 times.

#### b)

During the process, $p$ is evaluated $n$ times such that $\frac a {3^n} < 0.1$.

Solve it we have:

$n > \log_3 {10a}$

-> $n > \frac {\log {10a}} {\log 3}$

So, the number of steps for `sine` grows as $\Theta(\log a)$.

The interpreter must maintain the stack for that number of calls to p, so the space complexity is also $\Theta(\log a)$.

### 1.2.4  Exponentiation

````racket
(define (expt b n)
  (if (= n 0)
      1
      (* b (expt b (- n 1)))))
````

````racket
(define (expt b n)
  (expt-iter b n 1))

(define (expt-iter b counter product)
  (if (= counter 0)
      product
      (expt-iter b
                (- counter 1)
                (* b product)))) 
````

````racket
(define (fast-expt b n)
  (cond [(= n 0) 1]
        [(even? n) (square (fast-expt b (/ n 2)))]
        [else (* b (fast-expt b (- n 1)))]))
````

````racket
(define (even? n)
  (= (remainder n 2) 0))
````

#### Exercise 1.19

When apply $T_{pq}$ once:

$a = bq + aq + ap$

$b = bp + aq$

When apply $T_{pq}$ twice:

$$
\begin{flalign}
a &= (bp + aq)q + (bq + aq + ap)q + (bq + aq + ap)p && \\ \nonumber
  &= bpq + aq^2 + bq^2 + aq^2 + apq + bpq + apq + ap^2 && \\ \nonumber
  &= 2bpq + bq^2 + 2aq^2 + 2apq + ap^2 && \\ \nonumber
  &= b(q^2 + 2pq) + a(q^2 + 2pq) + a(p^2 + q^2) \nonumber
\end{flalign}
$$

$$
\begin{flalign}
b &= (bp + aq)p + (bq + aq + ap)q && \\ \nonumber
  &= bp^2 + apq + bq^2 + aq^2 + apq && \\ \nonumber
  &= bp^2 + bq^2 + 2apq + aq^2 && \\ \nonumber
  &= b(p^2 + q^2) + a(q^2 + 2pq) \nonumber
\end{flalign}
$$

So, let $p' = p^2 + q^2$ and $q' = q^2 + 2pq$, $T_{pq}^2 = T_{p'q'}$.

````racket
(define (fib n)
  (fib-iter 1 0 0 1 n))
(define (fib-iter a b p q count)
  (cond [(= count 0) b]
        [(even? count)
         (fib-iter a
                   b
                   (+ (square p) (square q))
                   (+ (square q) (* 2 p q))
                   (/ count 2))]
        [else (fib-iter (+ (* b q) (* a q) (* a p))
                        (+ (* b p) (* a q))
                        p
                        q
                        (- count 1))]))
````
